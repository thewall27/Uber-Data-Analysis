# UBER DATA ANALYTICS

<b>Dataset:</b>  From NYC Taxi & Limousine Commission (TLC) for April'14 in NYC
<br><b>Keywords:</b> Data Analysis, Data Visualisation
    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import os
```

```python
data = pd.read_csv(r"C:\Users\Dhaval\Projects\UberDataAnalytics\uber-raw-data-apr14.csv")
```

```python
data
```

```python
#Timestamp

data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
data.tail()
```

```python
#Function to get date of month

def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
data.tail()
```

```python
#Function to gte weekday 

def get_weekday(dt):
    return dt.weekday()
data['weekday'] = data['Date/Time'].map(get_weekday)

#function to get hours

def get_hour(dt):
    return dt.hour
data['hour'] = data['Date/Time'].map(get_hour)

data.tail()
```

<br> 
<b>Uber Trips w.r.t Day of month</b>

```python
plt.hist(data.dom, bins = 30, rwidth = 0.8, range=(0.5,30.5))
plt.xlabel('Date of Month')
plt.ylabel('Frequency')
plt.title('Number of Uber Trips w.r.t. DoM')
```

```python
def count_rows(rows):
    return len(rows)
by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
plt.bar(range(1, 31), by_date_sorted)
plt.xticks(range(1, 31), by_date_sorted.index)
plt.xlabel('Date of Month')
plt.ylabel('Frequency')
plt.title('Number of Uber Trips w.r.t. DoM')
```

<br>
<b>Uber Trip w.r.t. Hours</b>

```python
plt.hist(data.hour, bins=24, rwidth=0.8, range=(0.5,24))
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.title('Number of Uber Trips w.r.t. Hours')
```

<br>
<b>Uber Trips w.r.t. weekdays</b>

```python
plt.hist(data.weekday, bins=7, range=(-0.5,6.5), alpha=0.4, color='green', rwidth=0.8)
plt.xticks(range(7), 'Mon Tue Wed Thr Fri Sat Sun'.split())
plt.xlabel('Weekday')
plt.ylabel('Frequency')
plt.title('Number of Uber Trips w.r.t. weekdays')
```

<br>
<b>Cross Analysis of Uber Trips w.r.t Weekdays and Hours</b>

```python
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()
by_cross
```

```python
sb.heatmap(by_cross)
```
